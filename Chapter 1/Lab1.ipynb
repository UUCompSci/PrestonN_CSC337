```python
from cmath import nan
import pandas as pd
df = pd.read_excel('Film_Awards.xlsx', na_filter= True, header = None)
```

This will fill Na in the rows with blanks for a cleaner look

```python
df = df.fillna('')
```

I use the index function to drop any rows beginning with no values.

```python
df = df.drop(index=[5,12])
```

Change Olivia Coleman to correct name - "Olivia Colman" and 1917 to "1917" and fixes the \n on The favourite
Also switches columns 1,2 and 3,4 and 5,6 of best picture so that films are all in same row
Deletes Heading for every section and moves it to top.

```python
import numpy


array = df.to_numpy(str)
array[15,5] = "Olivia Colman"
array[2,3] = "1917"
array[15,6]= "The Favourite"
array[2:5, [2, 1]] = array[2:5, [1, 2]]
array[2:5, [4,3]] = array[2:5, [3,4]]
array[2:5, [6,5]] = array[2:5, [5,6]]
array = numpy.delete(array,1,0)
array = numpy.delete(array,6,0)
array = numpy.delete(array,11,0)
df1 = pd.DataFrame(array, columns = ['Year', 'Academy Award', 'Film','Golden Globe(Drama)','Film','Golden Globe(Musical or Comedy)','Film'])
df1
```

Splits the data into the Best Picture, Best Actor, and Best Actress sections

```python
picture = df1.iloc[0:4]
actor = df1.loc[5:8]
actress = df1.loc[10:13]
```

Convert Dataframes to arrays to more easily manipulate and delete rows and change column headers.

```python
import numpy

array = picture.to_numpy(str)
array = numpy.delete(array,5,1)
array = numpy.delete(array,3,1)
array = numpy.delete(array,1,1)
picture = pd.DataFrame(array, columns=['Year', 'Academy Award','Golden Globe(Drama)','Golden Globe(Musical or Comedy)'])

array = actor.to_numpy(str)
array = numpy.delete(array,0,1)
array = numpy.delete(array,0,0)
actor = pd.DataFrame(array, columns=[ "Actor1",'Academy Award',"Actor2",'Golden Globe(Drama)',"Actor3",'Golden Globe(Musical or Comedy)'])

array = actress.to_numpy(str)
array = numpy.delete(array,0,1)
array = numpy.delete(array,0,0)
actress = pd.DataFrame(array, columns=[ "Actor1",'Academy Award',"Actor2",'Golden Globe(Drama)',"Actor3",'Golden Globe(Musical or Comedy)'])
```

Melt picture so that each film is in row with its year and award

```python
picture = picture.melt(id_vars= "Year", value_vars=['Academy Award','Golden Globe(Drama)','Golden Globe(Musical or Comedy)'], value_name="Film" )
picture
```

Goes through column of films and returns rows of films that appear more than once.

```python
var = []
for item in picture["Film"]:
    if item in var:
        answerfilm = picture.loc[picture["Film"] == item]
    else:
        var.append(item)
answerfilm
```
