In the read excel function, I use "Index_col = 0" to get rid of the first column. I added the NA filter but the fillna function does its job anyway. I also get rid of the header to make it look cleaner. I can't however get rid of the numbers at the top. Im under the impression however, that those are removed in a finished product. 

```python
from cmath import nan
import pandas as pd
df = pd.read_excel('Film_Awards.xlsx', index_col = 0, na_filter= True, header = None)
```

This will fill Na in the rows with blanks for a cleaner look

```python
df = df.fillna('')
```

I use the index function to drop any rows beginning with no values and then print the results.

```python
df = df.drop(index=[nan])
df
```
